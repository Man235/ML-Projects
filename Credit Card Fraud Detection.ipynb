```python
#Importing dependencies
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
```

```python
# loading the dataset to a pandas dataframe
credit_card_data = pd.read_csv("creditcard.csv")
```

```python
credit_card_data.head()
```

```python
credit_card_data.tail()
```

```python
# Dataset information
credit_card_data.info()
```

```python
# Checking the number of missing values in each column
credit_card_data.isnull().sum()
```

```python
# Distribution of legit and fraudulent transaction
credit_card_data['Class'].value_counts()
```

```python
# This dataset is highly unbalanced
# 0 --> Legitimate Transaction
# 1 --> Fraudulent Transaction
```

```python
# Seperating the data for analysis
legit= credit_card_data[credit_card_data.Class==0]
fraud= credit_card_data[credit_card_data.Class==1]
```

```python
print(legit.shape)
print(fraud.shape)
```

```python
# Statistical measures of the data
legit.Amount.describe()
```

```python
fraud.Amount.describe()
```

```python
# Compare the values for both transactions
credit_card_data.groupby('Class').mean()
```

```python
# "Under-Sampling"
# Build a sample dataset containing similar distribution of normal transaction and fraudulent transaction
# Number of fraud transactions = 492
legit_sample = legit.sample(n=492)
```

```python
legit_sample.head()
```

```python
# Concatenating 2 dataframes
new_dataset= pd.concat([legit_sample, fraud], axis=0)
# axis=0 dataframe is added one by one row wise and axis=1 column wise
```

```python
new_dataset.head()
```

```python
new_dataset.tail()
```

```python
new_dataset['Class'].value_counts()
```

```python
# Now we have uniformly distributed data
```

```python
new_dataset.groupby('Class').mean()
```

```python
# Splitting the data into Feauture and Targets(either 0 or 1)
X = new_dataset.drop(columns='Class', axis=1)
Y = new_dataset['Class']
```

```python
X
```

```python
Y
```

```python
# Split the data into training data and testing data
X_train, X_test, Y_train, Y_test = train_test_split(X,Y, test_size=.2, stratify=Y, random_state=2)
```

```python
print(X.shape, X_train.shape, X_test.shape)
```

```python
# Model Training
# Logistic Regression

model = LogisticRegression(max_iter=1000)
```

```python
# Training the logistic regression model with training data
model.fit(X_train, Y_train)
```

```python
# Model Evaluation
# Accuracy Score
```

```python
# Accuracy on Training data
X_train_prediction = model.predict(X_train)
training_data_accuracy= accuracy_score(X_train_prediction, Y_train)
```

```python
print("Accuracy on training data : ", training_data_accuracy)
```

```python
# Accuracy on Test Data
X_test_prediction= model.predict(X_test)
test_data_accuracy= accuracy_score(X_test_prediction, Y_test)
```

```python
print('Accuracy score on test data : ', test_data_accuracy)
```
