```python
df1.info()
```

```python
y = df1['Target']
X = df1.drop(columns='Target',axis =1)

print(y.shape,X.shape)
y.value_counts()
```

```python
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.2,random_state=8997)

print(X_train.shape,X_test.shape,y_train.shape,y_test.shape)
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
X_train = pd.DataFrame(scaler.fit_transform(X_train), columns = X.columns)
X_test = pd.DataFrame(scaler.transform(X_test), columns = X.columns)
```

## LOGISTIC REGRESSION

```python
lr = LogisticRegression()
lr.fit(X_train, y_train)
```

```python
y_pred = lr.predict(X_test) ## Prdicting y values with the test data
```

```python
## Taining and Test Score
print("Model Score in Training data : ",lr.score(X_train, y_train)*100)
print("Model Score in Testing data : ",lr.score(X_test, y_test)*100)
```

```python
cf = confusion_matrix(y_test, y_pred)
cf
```

```python
from sklearn.metrics import accuracy_score, classification_report
print(classification_report(y_pred,y_test))
```

```python
print(accuracy_score(y_pred,y_test))
```
